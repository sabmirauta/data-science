```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
# Load the data from a .csv in the same folder. Since we will do some preprocessing, the variable is not called 'data' just yet!
raw_data = pd.read_csv('1.03. Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No':0})
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
```

```python
results.summary()
```

```python
plt.scatter(data['SAT'],y, c= data['Attendance'], cmap="RdYlGn_r")
yhat_yes = 0.8665 + 0.0014 * data['SAT']
yhat_no = 0.6439 + 0.0014 * data['SAT']
yhat = 0.0017 * data['SAT'] + 0.275
plt.plot(data['SAT'], yhat_no, lw=2, c="green")
plt.plot(data['SAT'], yhat_yes, lw=2, c="red")
plt.plot(data['SAT'], yhat, lw=2, c="blue")
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.show()
```
