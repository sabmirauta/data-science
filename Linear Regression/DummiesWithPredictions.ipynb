```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
plt.scatter(data['SAT'], y, c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014 * data['SAT']
yhat_yes = 0.8665 + 0.0014 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='green')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='red')
plt.xlabel('SAT', fontsize=18)
plt.ylabel('GPA', fontsize=18)
plt.show()
```

```python
plt.scatter(data['SAT'], y, c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014 * data['SAT']
yhat_yes = 0.8665 + 0.0014 * data['SAT']
yhat = 0.275 + 0.0017 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='green', label='regression_line_no')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='red', label='regression_line_yes')
fig = plt.plot(data['SAT'], yhat, lw=2, c='blue', label='regression_line')
plt.xlabel('SAT', fontsize=18)
plt.ylabel('GPA', fontsize=18)
plt.show()
```
