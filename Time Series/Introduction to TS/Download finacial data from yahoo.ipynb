```python
import yfinance as yf
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
raw_data = yf.download(tickers = "^GSPC ^FTSE ^N225 ^GDAXI", start="1994-01-07", end="2020-09-27", interval = "1d", group_by="ticker", auto_adjust=True, Threads=True)
```

```python
df_comp = raw_data.copy()
```

```python
df_comp.head()
```

```python
df_comp["spx"] = df_comp["^GSPC"].Close
df_comp["dax"] = df_comp["^GDAXI"].Close
df_comp["ftse"] = df_comp["^FTSE"].Close
df_comp["nikkei"] = df_comp["^N225"].Close
```

```python
df_comp = df_comp.iloc[1:]
```

```python
del df_comp["^GSPC"]
del df_comp["^GDAXI"]
del df_comp["^FTSE"]
del df_comp["^N225"]
```

```python
df_comp = df_comp.asfreq('b')
df_comp = df_comp.fillna(method = "ffill")
```
